works on monotonic fuction

Algo-

find mid elenment->
compare with key elenment->
perform parameter update and search until base case or until found

```python
def binary_serach(arr, elenment):
    start=0
    end=len(arr)-1
    mid=int((start+end)/2)
    while start<=end:
        if arr[mid]==elenment:
            return mid
        if elenment>arr[mid]:
            start=mid+1
        else:
            end=mid-1
        mid=int((start+end)/2)
    return -1
    
```

```python
arr=[2,4,6,9,10,13,15]
print(binary_serach(arr,2))
```

if the value of start and end are 2^31 - 1, then if we compute mid value it will overflow as the max value of integere is in 2^31-1,
to prevent this we can rewrite the alorigthm as follow

```python
def binary_serach(arr, elenment):
    start=0
    end=len(arr)-1
    mid=int(start + ((end-start)/2) )
    while start<=end:
        if arr[mid]==elenment:
            return mid
        if elenment>arr[mid]:
            start=mid+1
        else:
            end=mid-1
        mid=int(start + ((end-start)/2) )
    return -1
    
```

```python
arr=[2,4,6,9,10,13,15]
print(binary_serach(arr,2))

```
