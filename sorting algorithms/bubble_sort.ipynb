# bubble sort

algo 
- start with the first index -> compare the two adjacent elenemtn - > swap them if not in order
- repeat this process for n iterations

Tc- O(n^2)
Sc- O(1)

best case time complexity- O(n)
worst case time complexity- O(n^2)

```python
def bubble_sort(arr):
    for i in range(len(arr)-1, 1, -1):
        for j in range(0, i):
            if arr[j+1]<arr[j]:
                arr[j+1],arr[j] = arr[j], arr[j+1]
    return arr
    
```

```python
l=[1,3,7,3,1,5,6]
l=bubble_sort(l)
print(l)
```
