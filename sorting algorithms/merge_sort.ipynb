# merge sort



algo-> we break the array by finding mid till all the elenment are in single form-> merge all the arrays back to their original state by arranging in specific order

fast as compared to bubble seletion and other sorting alogrithm
'

Time compexity - O(n logn )
space complexity - O(n)


```python
def merge(arr, start, end):
    mid =start + (end-start)//2

    len1 = mid - start +1
    len2 = end - mid

    first = [0]*len1
    second = [0]*len2

    #copying the values
    k = start
    for i in range(len1):
        first[i]=arr[k]
        k+=1

    k=mid + 1
    for i in range(len2):
        second[i] = arr[k]
        k+=1

    #merging
    index1 = 0
    index2 = 0 
    k = start

    while(index1 < len1 and index2 < len2):
        if (first[index1] < second[index2]):
            arr[k] = first[index1]
            k,index1 = k+1, index1+1
        else:
            arr[k] = second[index2]
            k,index2 = k+1, index2+1

    while index1<len1:
        arr[k] = first[index1]
        k,index1 = k+1, index1+1

    while index2<len2:
        arr[k] = second[index2]
        k,index2 = k+1, index2+1

def merge_sort(arr, start, end):
    #base case
    if start>=end:
        return
    
    mid =start + (end-start)//2

    #sroting the left part
    merge_sort(arr, start, mid)

    #soring the right part
    merge_sort(arr, mid+1,end)

    #merge the arrays
    merge(arr, start, end)
```

```python
arr=[2,5,1,6,9]
arr2=[3,7,0,1,5,83,2,34,66,87,23,12,12,12]
merge_sort(arr2, 0 ,len(arr2)-1)
print(arr2)
```
