pick an elenment and put it on its right place - > now all the elenemtn smaller than this elenment will go on the left and the larger elenemnt will go on the right, this is know as partion karna -> now we use recursion to repeate the same process


Space- O(n)
Time - O(nlogn), worst case- o(n^2), best case- O(nlogn)

```python
"""

cosider arr=[3,1,4,5,2]
algo for partion -
get the pivot elenment(3 in this case) i.e the elenment that we want to place in the right palce -> count the number of elenemnt less than the pivot elenment( this is done so that we can 
place the pivot elenemnt on the right place) -> pivot will go to start + count index where count is the number of elenment less than pivot

NOW, we have to build the condition as ; a< | a | >a i.e elenment less than a on left and elenment greter than a on right -> we take to index i and j, i at left and j a
t right -> the value at i is 4 and at j is 2 i.e we swap the elenmtn as 4>3 and 2<4 -> now we add 1 to i and reduce j by 1, now value of elenment at i and j are 1 and 5 respectively,
 we increnment i and decrenment j and repeat the same process until i and j point to same index 

 After this my first pass is done of quick sort
 now we call quick sort for left and right half again


 """
def partion(arr, start, end):
    pivot = arr[start]

    #no of elenment less than pivot
    count = 0
    for i in range(start + 1, end):
        if arr[i]<=pivot:
            count+=1
    
    #palcing the pivot at right place, i.e swapping the elenments
    pivotIndex= start + count
    arr[start] , arr[pivotIndex]  =  arr[pivotIndex], arr[start] 
 

    #moving elenments less than a to left of a and elenment greter than a to right of a
    i,j=start, end
    while i<pivotIndex and j>pivotIndex:
        while arr[i] < arr[pivotIndex]:
            i+=1
        while arr[j] > arr[pivotIndex]:
            j-=1
        if i<pivotIndex and j>pivotIndex:
            #swaping i and j elnemnt
            arr[i] , arr[j] =  arr[j],arr[i] 
            i,j=i+1,j-1
            
    return pivotIndex
        

def quicksort(arr, start, end):
    if start>= end:
        return
    
    #performing the partion 
    p = partion(arr, start, end)
    #performing recursive call to sort left part and right part based on p 
    quicksort(arr, start, p-1)
    #performing recursive call to sort right part
    quicksort(arr, p+1, end)
```

```python
arr=[2,4,1,6,9,13,11,13,9,9]
quicksort(arr,0,len(arr)-1)
for k in arr:
    print(k)
```
