# selection sort

algo---
- traverse linearly
- for each traversal, replace the value at index i, put th minimum value of array
- repeat till end of line


Space complexity- O(1)
Time complexity- O(n^2)

best time complexity=same
worst time complexity=same

test case- use when array size is small or memory constraints


```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        min_index = i
        for j in range(i+1, len(arr)):
            if arr[j]<arr[min_index]:
                min_index=j 
        #swaping the values 
        arr[i],arr[min_index] = arr[min_index], arr[i]
    return arr
```

```python
l=[1,3,7,3,1,5,6]
l = selection_sort(l)
print(l)
```
